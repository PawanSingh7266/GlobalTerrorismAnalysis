## 7. Which country have able to successfull tackle (more than 5)  terrorist attacks ?

```python
# Chart - 7 visualization code
#stop attacks groupby and agg can help to make chart
stop_attack = df.groupby('country_txt').agg({'eventid':'count','success':'sum'}).reset_index()
stop_attack.columns = ['Country','Total_Attacks','Attack_Success']
#with the help of calculation we can find success and failure attacks
stop_attack['Attack_Failed'] = stop_attack['Total_Attacks']- stop_attack['Attack_Success']
stop_attack['Failure_Rate_Attack'] = (stop_attack['Attack_Failed']/ stop_attack['Total_Attacks'])*100
stop_attack['Success_Rate_Attack'] = (stop_attack['Attack_Success']/ stop_attack['Total_Attacks'])*100
stop_attack = stop_attack.round(2)
```

```python
#In chart we find those countries to able to stop attacks
stop_terror_attack = stop_attack[stop_attack['Total_Attacks']>=5].sort_values('Failure_Rate_Attack',ascending=False )[:10]
plot_able_stop_terror_attack = stop_terror_attack.loc[:,['Country','Failure_Rate_Attack']]
fx = plt.rcParams['figure.figsize'] = 16,8.27
fx = sns.barplot(x = 'Country',y = 'Failure_Rate_Attack',data = plot_able_stop_terror_attack)
#title can help to make a name of chart to anyone easily understand what chart says
fx.set_xlabel('Countries',fontsize = 18)
fx.set_ylabel('Rate of Failure of Terror Attack',fontsize = 18)
fx.set_title('Top 10 Countries Able to Succefull Tackle  Terrorist Attack("More Than 5 Attack")',fontsize=20);
```

### 1. Why did you pick the specific chart?

This line sets the title for the chart as 'Top 10 Countries Able to Successfully Tackle Terrorist Attacks. The specific chart, a bar plot, is chosen in this case because it allows for the comparison of the failure rates of terrorist attacks across different countries in a visually informative manner. The length of each bar represents the failure rate, and the chart enables easy identification of the top 10 countries with the highest failure rates in tackling such attacks.






### 2. What is/are the insight(s) found from the chart?

Identification of countries with the highest failure rates: The chart would allow you to easily identify the top 10 countries with the highest failure rates in handling terrorist attacks. These countries would have a comparatively lower failure rate, indicating successful counter-terrorism efforts. The chart could identify countries with higher failure rates as areas that may require further attention, resources, or changes in strategies to enhance their ability to prevent and mitigate terrorist attacks.
